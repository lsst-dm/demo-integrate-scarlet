# Summary of HSC Validation data prepared


We generated coadds with the w_2019_18 stack based on calexps, Sky corrections, and jointcal calibrations produced in Hsin-Fang's w_2019_18 monthly run in `/datasets/hsc/repo/rerun/RC/w_2019_18/DM-19151-jointcal`. We ran all 6 bands on tract=9813 (COSMOS) for its coverage with space-based data. 

Coadds were generated with 3 different skymaps with patch sizes:
 * default ("4k"): `config.skyMap.active.patchInnerDimensions=(4000, 4000)`
 * medium ("2k"): `config.skyMap.active.patchInnerDimensions=(2000, 2000)`
 * small ("1k"): `config.skyMap.active.patchInnerDimensions=(1000, 1000)`
 * tiny ("05k"): `config.skyMap.active.patchInnerDimensions=(500, 500)`
  
Coadds were generated at 2 depths:
 * default: all visits
     ```
     /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/w_2019_18_05k     
     /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/w_2019_18_1k
     /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/w_2019_18_2k
     /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/w_2019_18_4k```
     
 * shallow1: every other visit starting at 0 `visits[::2]`
     ```
    /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/shallow/w_2019_18_1k_1
    /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/shallow/w_2019_18_2k_1
    /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/shallow/w_2019_18_4k_1
    ```
    
 * shallow2: every other visit starting at 1 `visits[1::2]`
 
    ```
    /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/shallow/w_2019_18_4k_2
    /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/shallow/w_2019_18_1k_2
    /dataset/hsc/repo/rerun/private/yusra/deblender_sprint/shallow/w_2019_18_2k_2
    ```
